## 2 Theoretical notions

Linear prediction represents the estimation of a sample of the signal
$x[n]$ as a linear combination of $N$ previous samples:

$$x[n] \approx a_1 x[n-1] + a_2 x[n-2] + ... + a_N x[n-n]$$

Signals that approximately follow such a relationship are called
“autoregressive” (AR). $N$ represents the order of the autoregressive
model.

In Matlab, the function `lpc()` estimates the coefficients $a_k$ (refer
to the documentation).

An alternative and more accurate method is provided in the function
`lpc_exact()` along with the laboratory work.

## 3 Matlab examples

### Generating some simple signals

Linear signal:

``` matlab
x = linspace(0,20,50);
plot(x)
```

Sinusoidal signal:

``` matlab
n = 0:50;
f = 0.1;
x = sin(2*pi*f*n);
plot(x)
```

Exponential signal $a^n u[n]$:

``` matlab
n = 0:20;
x = (1/2).^n;
plot(x)
```

Noise:

``` matlab
x = randn(1,100);   % or use rand()
plot(x)
```

### Computing LPC coefficients

``` matlab
x = 1:1:10
order = 5;
a = lpc(x, order);
```

### Prediction using the LPC coefficients

``` matlab
x = 1:1:500
order = 5;
a = lpc(x, order);

% Predict value at n=11 and append to x:
n=501;
x(n) = sum( x(n-1:-1:n-order) .* (-a(2:end))   )
```
